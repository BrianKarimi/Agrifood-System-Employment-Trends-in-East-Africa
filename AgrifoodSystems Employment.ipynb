Convert Dataset from Wide to Long Format


```python
id_cols=[
    'Area Code', 'Area Code (M49)', 'Area', 'Source Code', 'Source',
    'Indicator Code', 'Indicator', 'Sex Code', 'Sex', 'Element Code',
    'Element', 'Unit'
]
```

```python
#Convert Year columns to long format

years_cols = [col for col in df.columns if col.startswith('Y')]

df_long = pd.melt (
    df,
    id_vars=id_cols,
    value_vars=years_cols,
    var_name= 'Year',
    value_name = 'Value'
)
df_long['Year']=df_long['Year'].str.replace('Y', '').astype(int)
df_long = df_long[
(df_long['Year'] >= 2000) &
(df_long['Year'] <= 2023)
].copy()
df_long.shape
```

```python
main_indicators=[ '21144', '21156', '21085']
```

```python
# Keep only the years we care about

# Focus on the three core indicators for trends
main_indicators = [21144, 21156, 21085]
trend_df = df_long[
    df_long['Indicator Code'].isin(main_indicators)].copy()


print("Shape of trend-focused data:", trend_df.shape)
print("Years range:", trend_df['Year'].min(), "–", trend_df['Year'].max())
print("\nCountries covered:", sorted(trend_df['Area'].unique()))
```

```python
east_africa = ['Kenya', 'Ethiopia', 'Uganda', 'United Republic of Tanzania', 'Rwanda']

employment_df= df_long[(df_long['Area'].isin(east_africa)) &
    (df_long['Indicator Code'] == 21144)].copy()
```

```python
#PLOT FIGURE

plt.figure(figsize=(10,5))

sns.lineplot(
    data= employment_df[employment_df['Sex'] == 'Total'],
    x='Year',
    y= 'Value',
    hue='Country',
    linewidth = 2.2,
    marker = 'o',
    markersize = 5,
    alpha = 0.9
)
plt.title("Employment in Agriculture (thousands of persons) – East Africa\nTotal (ILO modelled estimates(21144)) 2000–2023", fontsize=14, pad=15)
plt.ylabel("Employment (1,000 persons)")
plt.xlabel("Year")
plt.grid(True, alpha=0.25, linestyle='--')
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()
```

Objective 2: Employment Share %

```python
df_share_male = df_long[
    (df_long['Year'] >= 2000) &
    (df_long['Year'] <= 2023) &
    (df_long['Area'].isin(east_africa)) &
    (df_long['Indicator Code'] == 21156) &          # Share %
    (df_long['Sex'] == 'Male')
].copy()
country_map = {
    'United Republic of Tanzania': 'Tanzania',

}
df_share_male['Country'] = df_share_male['Area'].replace(country_map)

 # YoY Change_Pct
df_share_male = df_share_male.sort_values(['Country', 'Year'])
df_share_male['YoY_Change_pct'] = (df_share_male.groupby('Country')['Value'].pct_change() * 100).round(1)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# Top: Share %
sns.lineplot(data=df_share_male, x='Year', y='Value', hue='Country', 
             linewidth=2.3, marker='o', ax=ax1)
ax1.set_title("Male Share of Employment in Agriculture (%) – East Africa", fontsize=13)
ax1.set_ylabel("Share (%)")
ax1.grid(True, alpha=0.3)

# Bottom: YoY Change %
sns.lineplot(data=df_share_male, x='Year', y='YoY_Change_pct', hue='Country', 
             linewidth=2, linestyle='--', marker='s', ax=ax2, legend=False)
ax2.set_title("Year-on-Year Change in Share (%)", fontsize=13)
ax2.set_ylabel("YoY Change (%)")
ax2.set_xlabel("Year")
ax2.grid(True, alpha=0.3)

for ax in [ax1, ax2]:
    ax.set_xticks(range(2000, 2024, 1))
    ax.tick_params(axis='x', rotation=45, labelsize=9)

# One legend on top plot
ax1.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
linestyle = (0, (5, 5)), 
linewidth = 1.1,
alpha = 0.55,
marker = None,              
plt.tight_layout()
plt.show()
```

```python
df_share_female = df_long[
    (df_long['Year'] >= 2000) &
    (df_long['Year'] <= 2023) &
    (df_long['Area'].isin(east_africa)) &
    (df_long['Indicator Code'] == 21156) &          # Share %
    (df_long['Sex'] == 'Female')
].copy()
country_map = {
    'United Republic of Tanzania': 'Tanzania',

}
df_share_female['Country'] = df_share_female['Area'].replace(country_map)

 # YoY Change_Pct
df_share_female = df_share_female.sort_values(['Country', 'Year'])
df_share_female['YoY_Change_pct'] = (df_share_female.groupby('Country')['Value'].pct_change() * 100).round(1)
```

```python
east_africa = ['Kenya', 'Ethiopia', 'Uganda', 'United Republic of Tanzania', 'Rwanda', 'Burundi']

# Productivity (21085, Total only)
df_prod = df_long[
    (df_long['Year'] >= 2000) & (df_long['Year'] <= 2023) &
    (df_long['Area'].isin(east_africa)) &
    (df_long['Indicator Code'] == 21085) &
    (df_long['Sex'] == 'Total')
].copy()

# Shares % (21156, by gender)
df_shares = df_long[
    (df_long['Year'] >= 2000) & (df_long['Year'] <= 2023) &
    (df_long['Area'].isin(east_africa)) &
    (df_long['Indicator Code'] == 21156) &
    (df_long['Sex'].isin(['Male', 'Female']))
].copy()

# Shorter country names
country_map = {'United Republic of Tanzania': 'Tanzania'}
df_prod['Country'] = df_prod['Area'].replace(country_map)
df_shares['Country'] = df_shares['Area'].replace(country_map)

```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_prod, x='Year', y='Value', hue='Country', linewidth=2.2, marker='o')
plt.title('Agriculture Value Added per Worker (2015 US$) – East Africa\n(Total, 2000–2023)', fontsize=14)
plt.ylabel('Value Added per Worker (US$)')
plt.xlabel('Year')
plt.xticks(range(2000, 2024, 1), rotation=45, ha='right')
plt.grid(True, alpha=0.1)
plt.legend(title='Country', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
g = sns.relplot(
    data=df_shares,
    x='Year',
    y='Value',
    hue='Country',
    col='Sex',
    kind='line',
    linewidth=2,
    marker='o',
    height=5,
    aspect=1.3,
    col_order=['Male', 'Female']
)

g.set_titles(col_template="{col_name}")
g.set_axis_labels("Year", "Share of Employment in Agriculture (%)")
g.fig.suptitle("Gender Shares in Agricultural Employment – East Africa (2000–2023)", y=1.03, fontsize=14)
g.add_legend(bbox_to_anchor=(1.02, 0.8), title='Country')



plt.show()
```

```python
latest_year = df_shares['Year'].max()   # usually 2023 or 2022
df_latest = df_shares[df_shares['Year'] == latest_year]

plt.figure(figsize=(12, 7))
sns.barplot(
    data=df_latest,
    y='Country',
    x='Value',
    hue='Sex',
    palette={'Male': '#1f77b4', 'Female': '#ff7f0e'}
)

plt.title(f"Male vs Female Share in Agricultural Employment – East Africa\nLatest Year: {latest_year}", fontsize=14)
plt.xlabel("Share (%)")
plt.ylabel("Country")
plt.grid(True, axis='x', alpha=0.1)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

```python
# Age-related indicators
age_codes = [21066, 21088, 21089, 21090, 21086]

df_age = df_long[
    (df_long['Indicator Code'].isin(age_codes)) &
    (df_long['Area'].isin(east_africa)) &
    (df_long['Sex'].isin(['Male', 'Female'])) &
    (df_long['Value'].notna())
].copy()

# Add age group labels
age_labels = {
    21066: 'Total 15+',
    21088: '15–24 (Youth)',
    21089: '25–54 (Prime)',
    21090: '55–64',
    21086: '65+'
}
df_age['Age_Group'] = df_age['Indicator Code'].map(age_labels)

# Plot: Stacked bar - if data exists
if not df_age_latest.empty:
    plt.figure(figsize=(12, 7))
    sns.catplot(
        data=df_age_latest,
        x='Area',
        y='Value',
        hue='Age_Group',
        palette='viridis'
    )
    plt.title("Agricultural Employment by Age Group and Gender – East Africa\n(Latest available survey year per country)", fontsize=14)
    plt.ylabel("Employment (1,000 persons)")
    plt.xlabel("Country")
    plt.legend(title='Age Group', bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
